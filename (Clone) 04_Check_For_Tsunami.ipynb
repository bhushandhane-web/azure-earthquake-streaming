```python
import json

# 1. Run SQL to check for recent Tsunamis
# We check the last 20 minutes to match the ADF trigger schedule
query = """
    SELECT count(*) as count, first(place) as location 
    FROM databricks_cata.gold.gold_high_impact 
    WHERE tsunami_flag = 1 
      AND event_time > current_timestamp() - INTERVAL 20 MINUTES
"""
df = spark.sql(query)
row = df.first()

# 2. Prepare the Result for ADF
# We send back a JSON string: {"found": true, "location": "Japan"}
result = {
    "found": row['count'] > 0,
    "location": row['location'] if row['location'] else "Unknown Location"
}

print(f"Check complete: {result}")

# 3. Send Output to ADF
dbutils.notebook.exit(json.dumps(result))
```
